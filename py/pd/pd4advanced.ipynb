```python
from pydataset import data
mtcars= data('mtcars')
mtcars.head()
```

```python
mtcars['am'] = mtcars['am'].map({0: 'auto', 1: 'manual'})
mtcars.head()
```

```python
mtcars.index
```

```python
mtcars['car_name'] = mtcars.index
mtcars.head()
```
