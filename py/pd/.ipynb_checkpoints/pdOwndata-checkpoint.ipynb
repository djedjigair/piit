```python
import pandas as pd
import numpy as np
import random
```

```python
rollno = list(range(1, 21))
rollno
```

```python
sname = ('Alice', 'Tom', 'sophia', 'Ethian', 'Diana', 'Eric', 'Bob', 'Rihana', 'Bill', 'Katy',
'Max', 'Jennifer', 'Rosie', 'Alex', 'Ben', 'Emily', 'Sarah', 'Katrina', ' Quinn', 'kelly')
print(sname)
```

```python
genders_set = {"Male", "Female"}
gender= [random.choice(list(genders_set)) for _ in rollno]
print(list(gender))
```

```python
gender.count('Male')
```

```python
gender.count('Female')
```

```python
print('Males-', gender.count('Male') , 'Female-', gender.count('Female'))
```

```python
courses_dict = {1: "CS", 2: "IT", 3: "AI", 4: "DS"}
```

```python
course = [courses_dict[random.randint(1, 4)] for _ in rollno]
print(list(course))
```

```python
np.random.normal(70, 10, len(rollno))
```

```python
a= np.arange(10)
print(a)
np.clip(a, 2, 6)
```

```python
age= np.random.randint(18, 25, size=len(rollno))
print(age)
```

```python
print(rollno, '\n', sname, '\n', gender, '\n', course, '\n', age)
```

```python
df= pd.DataFrame({"rollno": rollno, "sname": sname, "gender":gender, "course": course, "age": age})
print(df)
```
