```python
import pandas as pd
```

```python
Customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']})

Rentals = pd.DataFrame({
    'RentalID': [101, 102, 103, 104, 105],
    'CustomerID': [1, 2, 2, 3, 5],
    'CarModel': ['Toyota', 'Honda', 'Ford', 'Tesla', 'BMW'],
    'DaysRented': [3, 5, 2, 4, 6],
    'CostPerDay': [40, 35, 70, 100, 90]})
```

```python
inner_join = pd.merge(Customers, Rentals, on='CustomerID', how='inner')
print(inner_join)
```

```python
left_join = pd.merge(Customers, Rentals, on='CustomerID', how='left')
print(left_join)
```

```python
right_join = pd.merge(Customers, Rentals, on='CustomerID', how='right')
print(right_join)
```

```python
full_join = pd.merge(Customers, Rentals, on='CustomerID', how='outer')
print(full_join)
```
